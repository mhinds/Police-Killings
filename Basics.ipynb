## Shootings by Race

```python
police_killings['raceethnicity'].value_counts().plot.bar(title='Police Killings in the US')
```

Above is the absolute values of shooting by race. What happens if I scale it by each race's percentage of total US population?

```python
scale = {'White':63.7,'Black':12.2,'Hispanic/Latino':16.3,'Unknown':2.2,'Asian/Pacific Islander':4.9,'Native American':0.7}
scale_series = pd.Series(scale,name='% Total US Pop')
racial_breakdown = pd.concat([police_killings['raceethnicity'].value_counts(),scale_series],axis=1)
racial_breakdown['scaled']=racial_breakdown['raceethnicity'] / racial_breakdown['% Total US Pop']
racial_breakdown['scaled'].plot.bar(title='Police Killings Scaled by % of Total US Population')
```

Clearly there is an overrepresentation of people who identify as black being shot by police. Native Americans and 'Unknown' are also overrepresented. 

People who identify as Hispanic or Latino are similarly represented to whites, while Asian and Pacific Islanders are least represented. 

## Income Distribution

While there are some outliers, the majority of people shot by police are on the lower end of the income distribution. This is true for individual and household income. The median personal income of those shot by police is \$22,348, but \$28,889 for the general US population. The median household income for those shot by police is \$42,759 compared to \$53,657 for the entire US. 


US general incomes from: http://www.deptofnumbers.com/income/us/

```python
fig = plt.figure(figsize=(20,5))
ax = []
income = police_killings[police_killings['p_income']!='-']['p_income'].astype(int)
h_income = police_killings['h_income'].dropna().astype(int)
ax.append(fig.add_subplot(1,2,1))
ax.append(fig.add_subplot(1,2,2))
income.plot.hist(bins=25,title='Personal Income Distribution of People Shot by Police',ax=ax[0])
h_income.plot.hist(bins=25,ax=ax[1],title='Household Income Distribution of People Shot by Police')
```

## Which State has the Most Police Shootings per Capita?
My home state of Oklahoma ranks number one in police killings. In general it seems southern and midwest states have more police shootings than states in the Northeast. Below I list the number of police killings per million people in each state. 

```python
state_pop = pd.read_csv('state_population.csv')
counts = police_killings['state_fp'].value_counts()
states = pd.DataFrame({"STATE": counts.index, "shootings": counts})
states = states.merge(state_pop,on='STATE')
states['rate'] = states['shootings']/(states['POPESTIMATE2015'] / 1000000)
states[['NAME','shootings','POPESTIMATE2015','rate']].sort_values(by='rate',ascending=False)
```

## What is the ratio of armed vs unarmed people by state?

Below I list the ratio of people killed by police in each state that are armed with a deadly weapon. This includes firearms, knives, and vehicles. For some states, there are only a handful of data points, and these ratios aren't very informative. Perhaps it would be better to break up the data more coarsely, say by region instead of state. In this case, I filtered out states with less than 6 killings. 

It would be interesting to cross correlate the armed ratio with the rankings here: http://gunlawscorecard.org/

```python
armed_counts = police_killings[(police_killings['armed'] == 'Firearm') | (police_killings['armed'] == 'Vehicle') | (police_killings['armed'] == 'Knife')]['state_fp'].value_counts()
armed_states = pd.DataFrame({"STATE": armed_counts.index, "shootings": armed_counts})
armed_states = armed_states.merge(state_pop,on='STATE')
armed_states[['NAME','shootings','POPESTIMATE2015']].sort_values(by='shootings',ascending=False)
states['armed_ratio'] = armed_states['shootings']/states['shootings']
states = states.fillna(0)
states[states['shootings'] > 5][['NAME','shootings','POPESTIMATE2015','rate','armed_ratio']].sort_values(by='armed_ratio',ascending=False)
```
